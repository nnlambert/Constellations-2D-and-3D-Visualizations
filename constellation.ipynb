## Project: Visualizing the Orion Constellation

In this project I am shifting into my study of visualization into the stars. Particularly, I will be utilizing the star Orion [Orion](https://en.wikipedia.org/wiki/Orion_(constellation)), a warrior God from ancient Greek mythology. 

I will be visualizing the Orion constellation in 3D using the Matplotlib function `.scatter()`. To learn more about the `.scatter()` you can see the Matplotlib documentation [here](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html). 

I will create a rotate-able visualization of the position of the Orion's stars to better deliver a sense of their actual positions. To achieve this, I will be mapping real data from outer space that maps the position of the stars in the sky.

The goal of this project is to understand spatial perspective. Once I visualize Orion in both 2D and 3D, I will be able to see the difference in the constellation shape humans see from earth versus the actual position of the stars that make up this constellation.

```python
#Importing necessary libraries
%matplotlib notebook
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Orion Coordinates in xyz planes
x = [-0.41, 0.57, 0.07, 0.00, -0.29, -0.32,-0.50,-0.23, -0.23]
y = [4.12, 7.71, 2.36, 9.10, 13.35, 8.13, 7.19, 13.25,13.43]
z = [2.06, 0.84, 1.56, 2.07, 2.36, 1.72, 0.66, 1.25,1.38]
```

```python
#plotting in 2d
plt.figure()
plt.subplot(1,1,1)
plt.scatter(x,y)
plt.show()
```

```python
#Plotting in 3d
fig_3d = plt.figure()
fig_3d.add_subplot(1,1,1,projection="3d")
constellation3d = plt.scatter(x,y,z)
constellation3d
```
